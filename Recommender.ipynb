```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

```python
#Data for Recommendation
DataForRecommendation = pd.read_excel("Champo.xlsx", sheet_name = 'Data for Recommendation')
```

```python
rec_data = DataForRecommendation.drop(['Unnamed: 21', 'Unnamed: 22', 'Customer .1',
       'H-2', 'P-5', 'M-1', 'A-9', 'C-2', 'JL', 'N-1', 'T-5', 'C-1', 'T-2',
       'I-2', 'PD', 'L-5', 'M-2', 'RC', 'P-4', 'T-4', 'PC', 'A-11', 'CC'], axis = 'columns')
```

```python
rec_data
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(rec_data.iloc[:, 1:])
```

```python
data_scaled
```

```python
# Create a user-item matrix
customers = rec_data["Customer "]
```

```python
carpet_attributes = rec_data.columns[1:]
```

```python
carpet_attributes
```

```python
user_item_matrix = pd.DataFrame(data_scaled, columns=carpet_attributes, index=customers)
```

```python
import tensorflow as tf

# Set the hyperparameters
num_hidden1 = 50
num_hidden2 = 10
num_epochs = 100
learning_rate = 0.001
```

```python
# Define the model architecture
input_layer = tf.keras.layers.Input(shape=(user_item_matrix.shape[1],))
hidden1 = tf.keras.layers.Dense(num_hidden1, activation='relu')(input_layer)
hidden2 = tf.keras.layers.Dense(num_hidden2, activation='relu')(hidden1)
output_layer = tf.keras.layers.Dense(user_item_matrix.shape[1], activation='sigmoid')(hidden2)
```

```python
# Create the model
model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```

```python
# Compile the model
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(lr=learning_rate))
```

```python
# Train the model
history = model.fit(user_item_matrix, user_item_matrix, epochs=num_epochs, verbose=0)
```

```python
# Make predictions for a particular customer
customer_name = 'H-2'
customer_data = rec_data.loc[rec_data['Customer '] == customer_name].iloc[:, 1:]
customer_predictions = model.predict(customer_data)
```

```python
# Get the recommended carpet attributes
recommendations = carpet_attributes[customer_predictions[0].argsort()[::-1]]
```

```python
# Create a DataFrame with two columns: Customer & Attribute 
df_recommendations = pd.DataFrame({'Attribute': recommendations, 'Customer': customer_name})

# Print the DataFrame
print(df_recommendations)
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(rec_data, test_size=0.2, random_state=30)
# Create user-item matrices for the training and test data
train_customers = train_data['Customer ']
train_user_item_matrix = pd.DataFrame(scaler.fit_transform(train_data.iloc[:, 1:]), columns=carpet_attributes, index=train_customers)

test_customers = test_data['Customer ']
test_user_item_matrix = pd.DataFrame(scaler.transform(test_data.iloc[:, 1:]), columns=carpet_attributes, index=test_customers)
```

```python
# Train the model on the training data
model.fit(train_user_item_matrix, train_user_item_matrix, epochs=num_epochs, verbose=0)
```

```python
# Make predictions for the test data
test_predictions = model.predict(test_user_item_matrix)

# Compute the MSE between the predicted and actual values
mse = np.mean(np.power(test_predictions - test_user_item_matrix.values, 2))
print(f'Mean squared error: {mse:.4f}')
```
